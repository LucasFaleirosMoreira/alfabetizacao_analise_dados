```python
import pandas as pd
import openpyxl
```

**Começando visualização e tratamento dos dados de municipios**

```python
#2023
df_resultados_metas2023 = pd.read_excel('./2023/DADOS/resultados_e_metas_municipios.xlsx')
df_alunos2023 = pd.read_csv('./2023/DADOS/TS_ALUNO.csv', sep=';', encoding='latin1')
df_estado2023 = pd.read_csv('./2023/DADOS/TS_ESTADO.csv', sep=';', encoding='latin1')
df_municipio2023 = pd.read_csv('./2023/DADOS/TS_MUNICIPIO.csv', sep=';', encoding='latin1')
```

```python
#2024
df_resultados_metas2024 = pd.read_excel('./2024/DADOS/resultados_e_metas_municipios_2024.xlsx')
df_alunos2024 = pd.read_csv('./2024/DADOS/TS_ALUNO.csv', sep=';', encoding='latin1')
df_estado2024 = pd.read_csv('./2024/DADOS/TS_ESTADO.csv', sep=';', encoding='latin1')
df_municipio2024 = pd.read_csv('./2024/DADOS/TS_MUNICIPIO.csv', sep=';', encoding='latin1')
```

Juntando as tabelas de anos diferentes em apenas uma tabela contendo todos os anos

```python
df_alunos = pd.concat([df_alunos2023, df_alunos2024], ignore_index=False)
df_estado = pd.concat([df_estado2023, df_estado2024], ignore_index=False)
df_municipio = pd.concat([df_municipio2023, df_municipio2024], ignore_index=False)
df_resultados_metas = pd.concat([df_resultados_metas2023, df_resultados_metas2024], ignore_index=False)
```

```python
df_alunos.columns = ['Ano', 'ID_UF', 'Sigla_UF', 'ID_Aluno', 'Serie', 'ID_Escola', 'Tipo_Escola', 'ID_Municipio', 'Nome_Municipio', 'Presenca', 'Preenchimento', 'ID_Caderno', 'Peso_Aluno', 'Proficiencia', 'Alfabetizado']
df_estado.columns = ['Ano', 'ID_UF', 'Sigla_UF', 'Serie', 'Tipo_rede', 'Pc_Aluno_Alfabetizado', 'Media_Nota_Avaliacao', 'PC_ALUNO_NIVEL_0_LP',	'PC_ALUNO_NIVEL_1_LP',	'PC_ALUNO_NIVEL_2_LP',	'PC_ALUNO_NIVEL_3_LP',	'PC_ALUNO_NIVEL_4_LP',	'PC_ALUNO_NIVEL_5_LP',	'PC_ALUNO_NIVEL_6_LP',	'PC_ALUNO_NIVEL_7_LP',	'PC_ALUNO_NIVEL_8_LP']
df_municipio.columns = ['Ano', 'ID_UF', 'Sigla_UF', 'ID_Municipio', 'Nome_Municipio', 'Serie', 'Tipo_rede', 'Pc_Aluno_Alfabetizado', 'Media_Nota_Avaliacao', 'PC_ALUNO_NIVEL_0_LP',	'PC_ALUNO_NIVEL_1_LP',	'PC_ALUNO_NIVEL_2_LP',	'PC_ALUNO_NIVEL_3_LP',	'PC_ALUNO_NIVEL_4_LP',	'PC_ALUNO_NIVEL_5_LP',	'PC_ALUNO_NIVEL_6_LP',	'PC_ALUNO_NIVEL_7_LP',	'PC_ALUNO_NIVEL_8_LP']
df_resultados_metas.columns = ['Ano', 'ID_UF', 'Sigla_UF', 'ID_Municipio', 'Nome_Municipio', 'Rede', 'Pc_Aluno_Alfabetizado', 'Meta_2024', 'Meta_2025', 'Meta_2026', 'Meta_2027', 'Meta_2028', 'Meta_2029', 'Meta_2030', 'NIVEIS_ALFABETIZACAO', 'Pc_Participacao', 'Pc_Alunos_Alfabetizados_2023', 'Pc_Alunos_Alfabetizados_2024']
```

```python
df_municipio = df_municipio.astype(str).apply(lambda x: x.str.replace('.', ',', regex=False))
df_estado = df_estado.astype(str).apply(lambda x: x.str.replace('.', ',', regex=False))
df_alunos = df_alunos.astype(str).apply(lambda x: x.str.replace('.', ',', regex=False))
df_resultados_metas = df_resultados_metas.astype(str).apply(lambda x: x.str.replace('.', ',', regex=False))
```

```python
df_alunos['AlfabetizadoINT'] = df_alunos['Alfabetizado'].astype(int)
```

```python
#teste de medidas

taxa_alfabetizados_uf = (
    df_alunos.groupby('Sigla_UF')['AlfabetizadoINT']
    .mean()
)
```

```python
taxa_alfabetizado_geral = (df_alunos['AlfabetizadoINT'] == 1).mean()
taxa_alfabetizado_geral
```
